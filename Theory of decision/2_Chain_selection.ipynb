# Описание критериев
Требования к цепочке:
   * Долговечность
   * Универсальность
   * Самодостаточность (чтобы ничего дополнительно не нужно было покупать)
   * Должна быть такой, чтобы все обзавидовались!

## Металл

Несмотря на то, что платина может согнуться, ее более трудно сломать. Прочность выше чем у золота, значит больше и долговечность, можно носить каждый день! Все будут смотреть и завидовать
$Оценка:10$

## Замок
Ставим оценки по надежности, мы же не хотим потерять цепочку
### Круглый замок
Круглый замок. Наиболее распространенный вид замка для цепочек и браслетов. Данный замковый механизм представляет собой кольцо с защелкой-пружиной. Круглый замок обладает достаточно высокой надежностью.
$Оценка:7$

### Замок-карабин
С виду напоминает часть петли и натянутую пружину, которая отведена в сторону. Очень надежный тип замка, поэтому подходит для массивных цепей.
$Оценка:9$

### Замок-коробка
С виду напоминает коробку со вставленной в нее и сложенной пополам планкой. Надежность такого замка зависит от качества его изготовления, от профессионализма мастера.
$Оценка:5$

### Магнитный замок
Данный вид замка представляет собой два магнита, которые декоративно оформлены в том же стиле, что и сама цепь. Принцип работы – магниты притягиваются друг к другу. Не самый надежный вариант для ювелирного изделия.
$Оценка:3$

### Замок-тогл
Данный вид застежки выглядит как кольцо и вертикальная перекладина, которая вставляется вовнутрь кольца, а после этого разворачивается поперек. Этот вид замка отлично подходит для бижутерии, так как его дизайн достаточно изыскан, но он совершенно ненадежен, поэтому не рекомендуется для цепочек из драгоценных металлов.
$Оценка: 1$

# Интегральная оценка только для золотых

```python
df_gold = df2.drop([4,5], axis = 0)
df_gold
```

```python
integral_cost = np.array([5,7,9,2])  #Веса: мощность, разгон, макс.скорость - 1, расход - 4, коробка передач - 6
line = np.array(df_gold.drop(['Articul','Cost', 'Metall'], axis = 1).loc[0:3])
df_gold['Rating'] = line.dot(integral_cost)
df_gold
```

```python
plt.scatter(df_gold['Cost'], df_gold['Rating'], marker='o')
ax = plt.gca()
for i in range(0,4):
    ax.text(df_gold.loc[i,'Cost'], df_gold.loc[i,'Rating'], df_gold.loc[i,'Articul'])
plt.title ('Rating')
```
